```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define my demand function
def q(x):
    return 20000 - 10000*x
```

```python
# define my profit function
def profit(x):
    return (x-0.6)*q(x)
```

```python
# feasible region 0.6 to 2
x=np.linspace(0.6,2,400)
```

```python
# Evaluation of the optimal price
px=profit(x)
```

```python
x_optimal=x[np.argmax(px)]
```

```python
px_optimal=px.max()
```

```python
print('Optimal Price : ',x_optimal)
print('Quantity sold : ',q(x_optimal))
print('Profit : ',px_optimal)
```

```python
plt.plot(x,px)
plt.axvline(x_optimal, linestyle="--")
plt.scatter([x_optimal], [px_optimal])
plt.xlabel('Price X (USD)')
plt.ylabel('Monthly Profit P(X) (USD)')
plt.title('Profit Maximization')
plt.show()
```
