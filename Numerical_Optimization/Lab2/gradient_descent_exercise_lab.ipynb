## Part B — Problem Formulation

We model waiting time with **linear regression**:
$$
\hat y = X\theta
$$
where:
- $X\in\mathbb{R}^{m\times d}$ (features)
- $\theta\in\mathbb{R}^{d}$ (parameters)
- $y\in\mathbb{R}^{m}$ (target)

### Q4. Objective function
Write the Mean Squared Error objective:
$$
J(\theta) = \frac{1}{2m}\sum_{i=1}^m (\hat y_i - y_i)^2
$$

### Q5. Gradient
Derive the gradient:
$$
\nabla J(\theta) = \frac{1}{m}X^\top(X\theta - y)
$$

### Q6. Gradient Descent update rule
$$
\theta_{k+1} = \theta_k - \alpha \nabla J(\theta_k)
$$
Explain what \(\alpha\) (learning rate) does.
